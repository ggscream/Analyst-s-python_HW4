```python
# Условие 1.1 Скачать данные 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('./kc_house_data.csv')
df
```

```python
# Условие 1.2 Изучите стоимости недвижимости
plt.figure(figsize=(6,4))
sns.boxplot(data=df['price']);
```

```python
print(f"Процент домов, стоимостью меньше 1.5 миллионов = {(df['price'] < 2500000).sum()/df['id'].count():%}")
```

```python
plt.hist(df['price'], bins=100)
plt.xlim(0, 2500000)
plt.title('Цены домов')
plt.xlabel('Цена')
plt.ylabel('Количество');
```

```python
# Условие 1.3 Изучите распределение квадратуры жилой площади
plt.figure(figsize=(6,4))
sns.boxplot(data=df['sqft_living']);
```

```python
print(f"Процент домов с жилой площадью между 500 и 6000 = {((df['sqft_living'] < 6000)&(df['sqft_living'] > 500)).sum()/df['id'].count():%}")
```

```python
plt.hist(df['sqft_living'], bins=100)
plt.xlim(500, 6000)
plt.title('Жилая площадь')
plt.xlabel('Площадь')
plt.ylabel('Количество');
```

```python
# Условие 1.4 зучите распределение года постройки
plt.figure(figsize=(6,4))
sns.boxplot(data=df['yr_built']);
```

```python
plt.hist(df['yr_built'], bins=100)
plt.title('Годы постройки')
plt.xlabel('Год')
plt.ylabel('Количество');
```

```python
# Условие 2.1 Изучите распределение домов от наличия вида на набережную
data = df['waterfront'].value_counts()
data.index = ['Нет', 'Да']

plt.figure(figsize=(6, 4))
plt.pie(data, autopct='%1.1f%%')
plt.title('Дома с видом на набережную')
plt.legend(data.index);
```

Видом на набережную обладают только 0.8 % домов. Ввиду минимального количества домов с видом на набережную оценить влияние этого параметра на стоимость и другие характеристики затруднидельно.

```python
# Условие 2.2 Изучите распределение этажей домов
counts = df['floors'].value_counts() 
plt.figure(figsize=(8, 6))    
plt.title('Количество домов по этажам')
sns.barplot(x=counts.index, y=counts.values) 
plt.xlabel('Этажи')
plt.ylabel('Количество');
```

Подавляющее большинство домов имеют один или два этажа. Влияние этажности на стоимость домов будет рассмотрено в задаче 3.

```python
# Условие 2.3 Изучите распределение состояния домов
counts = df['condition'].value_counts() 
plt.figure(figsize=(8, 6))    
plt.title('Количество домов в разном состоянии')
sns.barplot(x=counts.index, y=counts.values) 
plt.xlabel('Состояние')
plt.ylabel('Количество');
```

Подавляющее большинство домов имеют состояние 3 или 4. Количество домов, имеющих оценку 1 или 2 несущественно. Влияние оценки на стоимость домов будет рассмотрено в задаче 3.
